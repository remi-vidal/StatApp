```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df_questions = pd.read_csv("../raw_data/deep_course_formulaire_question.csv")
```

```python
#garde que les questions indiquées par Simon
liste=["F7664060CC52BC6F3D620BCEDC94A4B6","9A1158154DFA42CADDBD0694A4E9BDC8","AC1DD209CBCC5E5D1C6E28598E8CBBE8",
                "289DFF07669D7A23DE0EF88D2F7129E7","19F3CD308F1455B3FA09A282E0D496F4"]
df_questions_retenues=df_questions[df_questions["id"].isin(liste)]
```

```python
#Suppression de colonnes inutiles
df_questions_retenues = df_questions_retenues.drop(labels = ["aide", "enonce_groupe", "type", "unite"], axis = 1)
df_questions_retenues.rename(columns = {'id': 'question_id'},inplace=True)
df_questions_retenues
```

```python
df_reponses = pd.read_csv("../raw_data/deep_course_formulaire_reponse.csv")
```

```python
#Base des réponses des clients pour les réponses retenues
df_reponses_retenues = df_reponses.loc[df_reponses['question_id'].isin(df_questions_retenues["question_id"])]
```

```python
df_reponses_retenues = df_reponses_retenues.dropna(subset = ["valeur_actee"]).reset_index(drop=True)
df_reponses_retenues = df_reponses_retenues.drop(labels = ["non_concerne_acte", "commentaire_acte"], axis = 1)
df_reponses_retenues["valeur_actee"] = df_reponses_retenues["valeur_actee"].apply(lambda x : x.replace("-",""))  #on élimine les tirets
df_reponses_retenues["valeur_actee"] = df_reponses_retenues["valeur_actee"].str.upper() #On met en majuscules pour coincïder avec df_choix
```

```python
df_reponses_retenues
```

```python
#Choix possible des réponses pour les questions retenues
df_choix = pd.read_csv("../raw_data/deep_course_formulaire_question_choix.csv", low_memory = False)
df_choix = df_choix.loc[df_choix['question_id'].isin(df_questions_retenues["question_id"])]
df_choix
```

```python
#réponses (avec 1 ou 0) des clients aux questions 
df_rep = pd.merge(df_reponses_retenues, df_choix, how='left', left_on=['valeur_actee'], right_on=['choix_id'])
df_rep
```

```python
df_rep = df_rep.drop(labels = ["valeur_actee", "question_id_x","choix_id"], axis = 1)
```

```python
#Garde que les réponses "Oui" (du moins les 1) 
#Autre base mais avec les 0
indexNames1 = df_rep[ df_rep['choix_order'] == 0 ].index
indexNames0 = df_rep[ df_rep['choix_order'] == 1 ].index
df_rep0 = df_rep.drop(indexNames0)
df_rep.drop(indexNames1 , inplace=True)
df_rep.rename(columns = {'question_id_y': 'question_id'},inplace=True)
df_rep0.rename(columns = {'question_id_y': 'question_id'},inplace=True)
df_rep
```

```python
#Mets les clients dans une liste en fonction de leurs réponses aux questions retenues
list_clients_gazo = df_rep[ df_rep['question_id'] == "F7664060CC52BC6F3D620BCEDC94A4B6" ]
list_clients_gazo = list(list_clients_gazo["node_id"])

list_clients_nanomateriaux = df_rep[ df_rep['question_id'] == "9A1158154DFA42CADDBD0694A4E9BDC8" ]
list_clients_nanomateriaux = list(list_clients_nanomateriaux["node_id"])

list_clients_radioprotection = df_rep[ df_rep['question_id'] == "AC1DD209CBCC5E5D1C6E28598E8CBBE8" ]
list_clients_radioprotection = list(list_clients_radioprotection["node_id"])

list_clients_equipements = df_rep[ df_rep['question_id'] == "289DFF07669D7A23DE0EF88D2F7129E7" ]
list_clients_equipements = list(list_clients_equipements["node_id"])

list_clients_GES = df_rep[ df_rep['question_id'] == "19F3CD308F1455B3FA09A282E0D496F4" ]
list_clients_GES = list(list_clients_GES["node_id"])
```

```python
#Pareil mais pour les réponses 0
list_clients_gazo0 = df_rep0[ df_rep0['question_id'] == "F7664060CC52BC6F3D620BCEDC94A4B6" ]
list_clients_gazo0 = list(list_clients_gazo0["node_id"])

list_clients_nanomateriaux0 = df_rep0[ df_rep0['question_id'] == "9A1158154DFA42CADDBD0694A4E9BDC8" ]
list_clients_nanomateriaux0 = list(list_clients_nanomateriaux0["node_id"])

list_clients_radioprotection0 = df_rep0[ df_rep0['question_id'] == "AC1DD209CBCC5E5D1C6E28598E8CBBE8" ]
list_clients_radioprotection0 = list(list_clients_radioprotection0["node_id"])

list_clients_equipements0 = df_rep0[ df_rep0['question_id'] == "289DFF07669D7A23DE0EF88D2F7129E7" ]
list_clients_equipements0 = list(list_clients_equipements0["node_id"])

list_clients_GES0 = df_rep0[ df_rep0['question_id'] == "19F3CD308F1455B3FA09A282E0D496F4" ]
list_clients_GES0 = list(list_clients_GES0["node_id"])
```

```python
#Création d'une DB avec juste les clients et les textes qui leur sont attribués (groupe ID)
df_text_versions_node = pd.read_csv("../raw_data/deep_course_cf_text_version_nodes.csv", low_memory = False)
df_text_versions = pd.read_csv("../raw_data/deep_course_cf_text_versions.csv", low_memory = False)
df_text_versions = df_text_versions.drop(labels = ["txt_name", "theme_id","type_id"], axis = 1)
df_text = pd.merge(df_text_versions_node, df_text_versions, how='left', left_on=['txt_version_surrogate_uuid'], right_on=['surrogate_uuid'])
df_text = df_text.drop(labels = ["txt_version_surrogate_uuid", "surrogate_uuid"], axis=1)
df_text
```

```python
#définission d'une fonction qui va prendre en compter l'occurrence d'un texte dans un tableau

def count(df,txt):
    c=0
    for i in np.arange(df.shape[0]) :
        if df.iloc[i,1]==txt :
            c=c+1
    return c
```
