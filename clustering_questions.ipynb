## Sélection des questions pertinentes

```python
df_questions = pd.read_csv("./data/deep_course_formulaire_question.csv", low_memory = False)
df_questions
```

On ne vas garder que les questions fermées : type 1 (yes no question) et 2 (QCM avec choix cumulatifs) :

```python
df_questions = df_questions.loc[df_questions["type"].isin([1,2])].drop(columns = ["enonce_groupe", "unite", "aide"])
df_questions
```

```python
# On élimine les questions qui étaient pour les groupes uniquement : 
df_questions = df_questions[df_questions['enonce_site'].notna()].reset_index(drop=True)
df_questions
```

Virer les réponses des comptes tests. Virer les questionnaires avec pas assez de réponses

## Etude des réponses

```python
df_choix = pd.read_csv("./data/deep_course_formulaire_question_choix.csv", low_memory = False)
df_choix
```

```python
df_reponses = pd.read_csv("./data/deep_course_formulaire_reponse.csv", low_memory = False)
df_reponses
```

```python
# Sélection des réponses aux questions qui nous intéressent
df_reponses = df_reponses.loc[df_reponses['question_id'].isin(df_questions["id"])]
```

```python
df_reponses
```

```python
#On élimine les lignes ou la valeur actée n'est pas remplie
df_reponses = df_reponses.dropna(subset = ["valeur_actee"]).reset_index(drop=True)
df_reponses
```

### Distribution du nombres de questions répondues

```python
df_reponses.groupby("node_id")["node_id"].count().plot.hist(bins=20)   
#Eliminer les questionnaires avec moins de 100 réponses ?
```

```python
# Elimination des clients n'ayant pas répondu au moins "rep_min" questions

rep_min = 100 
v = df_reponses.node_id.value_counts()
df_reponses[df_reponses.node_id.isin(v.index[v.gt(rep_min)])]
```
